```python
# Discrepancy checks and exploratory data analysis for the life expectancy datasets(male, female, both) from UN Data
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("seaborn-v0_8")
sns.set()
```

```python
male = pd.read_csv("../data_clean/life_expectancy_male_clean.csv")
female = pd.read_csv("../data_clean/life_expectancy_female_clean.csv")
both = pd.read_csv("../data_clean/life_expectancy_both_clean.csv")

male.head()
```

```python
print("Male:")
display(male.head())

print("Female:")
display(female.head())

print("Both sexes:")
display(both.head())
```

```python
#Discrepancy checks

print("=== MALE INFO ===")
male.info()
print("\n=== FEMALE INFO ===")
female.info()
("\n=== BOTH INFO ===")
both.info()
```

```python
print("=== Missing values (male) ===")
print(male.isna().sum())

print("\n=== Missing values (female) ===")
print(female.isna().sum())

print("\n=== Missing values (both) ===")
print(both.isna().sum())
```

```python
print("=== Male describe ===")
display(male.describe())

print("\n=== Female describe ===")
display(female.describe())

print("\n=== Both describe ===")
display(both.describe())
```

```python
print("Duplicates in male:", male.duplicated().sum())
print("Duplicates in female:", female.duplicated().sum())
print("Duplicates in both:", both.duplicated().sum())
```

```python
male = male.drop_duplicates()
female = female.drop_duplicates()
both = both.drop_duplicates()
```

```python
print("Male life expectancy min/max:",
male["LifeExpectancy"].min(),
male["LifeExpectancy"].max())
print("Female life expectancy min/max:",
female["LifeExpectancy"].min(),
female["LifeExpectancy"].max())
print("Both life expectancy min/max:",
both["LifeExpectancy"].min(),
both["LifeExpectancy"].max())

print("Year range (male):",
male["Year"].min(), "-", male["Year"].max())
print("Year range (female):",
female["Year"].min(), "-", female["Year"].max())
print("Year range (both):",
both["Year"].min(), "-", both["Year"].max())
```

```python
#Discrepancies
#Incorrect Data Types: Year was stored as string
#Duplicates: Several duplicate rows were found, were detected using df.duplicated() and removed
#Non-country entries: aggregated regions, for e.g ABD region: Central and West Africa distort country-level comparisons.
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(data=both, x="LifeExpectancy",
bins=30, kde=True)
plt.title("Distribution of Life Expectancy (both sexes, all countries, all years)")
plt.xlabel("Life Expectancy (years)")
plt.ylabel("Count")
plt.show()
```

```python
life_mf = male.merge(
    female,
    on=["Country", "Year"],
    suffixes=("_male", "_female")
)

life_mf.head()
```

```python
plt.figure(figsize=(6, 6))
sns.scatterplot(
    data=life_mf,
    x="LifeExpectancy_male",
    y="LifeExpectancy_female",
    alpha=0.4
)
plt.title("Male vs Female Life Expectancy (per Country-Year)")
plt.xlabel("Male life expectancy")
plt.ylabel("Female life expectancy")
plt.axline((0, 0), slope=1, color="red", linestyle="--")
plt.show()
```

```python
life_all = life_mf.merge(
    both,
    on=["Country", "Year"],
    how="inner"
)

life_all = life_all.rename(columns={"LifeExpectancy": "LifeExpectancy_both"})
life_all[["LifeExpectancy_male", "LifeExpectancy_female", "LifeExpectancy_both"]].head()
```

```python
plt.figure(figsize=(10,6))

sns.histplot(life_all['LifeExpectancy_male'],
kde=True, label="Male")

sns.histplot(life_all['LifeExpectancy_female'],
kde=True, color='orange', label="Female")

sns.histplot(life_all['LifeExpectancy_both'],
kde=True, color="green", label="Both")

plt.xlabel("Life Expectancy")
plt.legend()
plt.title("Distribution of Life Expectancy")
plt.show()
```
