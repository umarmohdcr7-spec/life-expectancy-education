```python
import pandas as pd

file_path = "data_raw/primary.csv"   # NOT ../ !!
datafile = pd.read_csv(file_path)

datafile.head()
```

```python
datafile = datafile.rename(columns = {
    "Reference Area" : "Country",
    "Time " : "Year",
    "Sex" : "Gender",
    "Observation Value" : "Number "
})

datafile.head()
```

```python
def clean_life_expectancy_un(input_path, output_path):
    df = pd.read_csv(input_path)
    df = df[df["Variant"] == "Medium"].copy()
    df = df[["Country or Area", "Year(s)", "Value"]].copy()

    df = df.rename(columns={
        "Country or Area": "Country",
        "Year(s)": "Year",
        "Value": "LifeExpectancy"
    })

    df["Year"] = df["Year"].astype(int)
    df["LifeExpectancy"] = pd.to_numeric(df["LifeExpectancy"], errors="coerce")

    df = df.dropna(subset=["LifeExpectancy"]).reset_index(drop=True)

    df.to_csv(output_path, index=False)
    return df.head()
```

```python
clean_life_expectancy_un(
    "data_raw/life_expectancy_male.csv",
    "data_clean/life_expectancy_male_clean.csv"
    )
```
